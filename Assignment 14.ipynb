# Question 5 Number Analyzer

Maya is creating a program to analyze numbers for her math class. She wants to create a function that takes a positive integer and provides different information about it based on its characteristics.

Place value reminder:
- The ones digit is the rightmost digit.
- The tens digit is immediately to the left of the ones digit.
- The hundreds digit is immediately to the left of the tens digit, and so on.
  For example, in 237: hundreds = 2, tens = 3, ones = 7.

Input validation:
- The input must be a positive integer. If it is not, print a clear message (for example: "Invalid input: please enter a positive integer") and stop.

The program should work as follows:

**If the number is less than 10:**
- Print "Single digit"
- Print the ones digit (which is the number itself)

**If the number is between 10 and 99 (inclusive):**
- Print "Double digit"
- Compute and print the result of dividing the ones digit by the tens digit using integer division
- Compute and print the remainder when the ones digit is divided by the tens digit

**If the number is between 100 and 999 (inclusive):**
- Print "Three-digit number"
- Calculate and print the sum of all digits in the number
- Round the number to 1 significant figure and print the result (nearest hundred)
- Round the number to 2 significant figures and print the result (nearest ten)

**If the number is 1000 or greater:**
- Print "Four or more digits"
- Calculate and print the sum of all digits in the number
- Calculate and print how many digits the number has

Guidance: Rounding to significant figures
- Find the first non-zero digit by scanning from left to right (from the most significant digit).
- To round to 1 significant figure: keep only the first digit; look at the next digit to decide whether to round up (5–9) or stay (0–4); replace all later digits with zeros.
- To round to 2 significant figures: keep the first two digits; look at the third digit to decide whether to round up or stay; replace the remaining digits with zeros.
- For three-digit numbers, 1 significant figure is the nearest hundred; 2 significant figures is the nearest ten.
- For this question, numbers with four or more digits do not require rounding; just report the sum of digits and the number of digits.

For example:
- Input: `7` should output:
  ```
  Single digit
  7
  ```
- Input: `45` should output:
  ```
  Double digit
  1
  1
  ```
  (Because ones = 5 and tens = 4; 5 divided by 4 is 1 with remainder 1)
  
- Input: `237` should output:
  ```
  Three-digit number
  12
  200
  240
  ```
  (Because 2+3+7 = 12; to 1 significant figure 237 rounds to 200; to 2 significant figures 237 rounds to 240)

- Input: `7654` should output:
  ```
  Four or more digits
  22
  4
  ```
  (Because 7+6+5+4 = 22; the number has 4 digits)
